## Orders of growth --> this approach used in industries to check the algorithm efficiency

- We want to evaluate program's efficiency when input is very big
- We want to express the growth of program's run time as input size grows
- We want to put an upper bound on growth --> as tight as possible
- Do not need to be precise : 'order-of' not 'exact' growth
- We will look at the largest factors in run time (which section of the program will take the longest to run)

```python
n=int(input('Enter the number to calculate factorial:'))
answer=1  # 1 operation
while n>1: # 1 operation
    answer*=n # 2 operation 
    n-=1 # 2 operation
return answer #1 operation

# Total opearation is 5n+2 hence time complexity is O(n)
```

Q.10 Reverse the list --> Time complexity is O(n) since the loop is going to n/2 and 2 is multiplicative constant so i will remove it thus time complexity is O(n)

```python
L=[1,2,3,4,5]
for i in range(0, len(L)//2):
    other=len(L)-i-1
    temp=L[i]
    L[i]=L[other]
    L[other]=temp
    
print(L)
```
